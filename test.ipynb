```python
import pandas as pd
import numpy as np
import plotly.express as px

pd.options.plotting.backend = "plotly"
```

```python
period_length = 500
price = pd.Series(
    data=(1 + np.random.normal(0, 0.01, period_length)).cumprod() * 100,
    index=pd.date_range(
        end=pd.Timestamp.today(), periods=period_length, normalize=True
    ).rename("date"),
)
window = 20
ew_mean = price.ewm(span=window, min_periods=window).mean()
ew_std = price.ewm(span=window, min_periods=window).std()
upper_band = price + 2 * ew_std
lower_band = price - 2 * ew_std
mean_dist = price - ew_mean
zscore = mean_dist / ew_std

display(
    pd.DataFrame(
        {
            "price": price,
            "ew_mean": ew_mean,
            "upper_band": upper_band,
            "lower_band": lower_band,
        }
    ).plot(height=300)
)


display(
    pd.DataFrame(
        {
            "zscore": zscore,
            "ew_std": ew_std,
            "mean_dist": mean_dist,
        }
    ).plot(height=300)
)
```
